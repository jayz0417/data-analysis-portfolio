3. 상관 관계 분석
- 변수 간의 상관 관계를 분석   
  - 비수치형 데이터를 인코딩:  boolean 타입의 데이터를 정수형으로 변환하고, 범주형 데이터에 대해서는 레이블 인코딩

1. 특정 변수로 축소:
- 변수가 너무 많아, 영향도가 높은 변수로 드릴다운 진행
- 관계가 높은 변수를 상위 10개로 축소

2. 주요 상관관계도:
- 현재 직무기간: 근속년수(+0.7), 현재 상사와 근무기간(+0.7)   
- 스톡옵션정도: 기혼여부(-0.7)    
- 연봉인상률: 업무성과도(+0.8)   
- 연령: 월소득(+0.5), 총근속년수(+0.7), 업무강도(+0.5)   

````python
from sklearn.preprocessing import LabelEncoder

# Boolean 타입을 int 타입으로 변환
bool_columns = df.select_dtypes(include=['boolean']).columns
df[bool_columns] = df[bool_columns].astype(int)

# 레이블 인코더 생성
label_encoder = LabelEncoder()

# 범주형 데이터에 레이블 인코딩 적용
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

# 상관관계 계산
corr_matrix = df.corr()

# 상관관계 시각화를 위한 Plotly Express 사용
import plotly.express as px

fig = px.imshow(corr_matrix,
                text_auto=True,
                aspect="auto",
                labels=dict(x="Feature", y="Feature", color="Correlation"),
                x=corr_matrix.columns,
                y=corr_matrix.columns,
                title="Correlation Matrix of HR Dataset")
fig.show()
````

````python
# 상관관계가 높은 변수들만 필터링 (절대값 0.5 이상)
high_corr_var = np.where(abs(corr_matrix) > 0.5)

# 필터링된 변수들의 쌍을 저장
high_corr_pairs = [(corr_matrix.columns[x], corr_matrix.columns[y]) for x, y in zip(*high_corr_var) if x != y and x < y]

# 필터링된 변수들의 유니크한 이름 리스트 생성
unique_high_corr_vars = list(set([item for sublist in high_corr_pairs for item in sublist]))

# 필터링된 변수들만 포함하는 새로운 데이터프레임 생성
filtered_df = df[unique_high_corr_vars]

# 새로운 데이터프레임의 상관관계 계산
filtered_corr_matrix = filtered_df.corr()

# 상관관계 시각화
fig = px.imshow(filtered_corr_matrix,
                text_auto=True,
                aspect="auto",
                labels=dict(x="Feature", y="Feature", color="Correlation"),
                x=filtered_corr_matrix.columns,
                y=filtered_corr_matrix.columns,
                title="Filtered Correlation Matrix of HR Dataset")
fig.show()
````

#### 가설수립
- 사회 초년생일 수록, 회사에 합류한지 얼마안될 수록, 현재 상사와 일한 기간이 짧을 수록 퇴사가 많은 건 아닐까   
  - TotalWorkingYears, YearsInCurrentRole, YearsWithCurrManager, YearsAtCompany와 같은 근속 연수 관련 변수들이 퇴직(Attrition)에 미치는 영향을 분석

````python
# 근속 연수 관련 변수 리스트
working_years_vars = ['TotalWorkingYears', 'YearsInCurrentRole', 'YearsWithCurrManager', 'YearsAtCompany']

# 각 변수에 대해 구간 분할 및 퇴직 비율 계산 후 시각화
for var in working_years_vars:
    # 구간 분할
    df[var + '_bin'] = pd.cut(df[var], bins=[-1, 5, 10, 15, 20, 25, 30, 35, 40], labels=['0-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40'])

    # 구간 별 퇴직 비율 계산
    attrition_rate_by_bin = df.groupby(var + '_bin')['Attrition'].mean() * 100

    # 시각화
    fig = px.bar(attrition_rate_by_bin, title=f'Attrition Rate by {var} Bins', labels={'value': 'Attrition Rate (%)', 'index': f'{var} Bins'})
    fig.update_layout(xaxis_title=f"{var} Bins", yaxis_title="Attrition Rate (%)")
    fig.show()
````

1. TotalWorkingYears, YearsInCurrentRole, YearsWithCurrManager, YearsAtCompany와 같은 근속 연수 관련 요인별로 퇴사와 관련 높은 상위 10개 요인간의 관계 분석

````python
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler

# 정규화를 위한 MinMaxScaler 인스턴스 생성
scaler = MinMaxScaler()

# 상위 10개 요인 + 'Attrition'
factors = [
    'OverTime', 'TotalWorkingYears', 'JobLevel', 'MaritalStatus',
    'YearsInCurrentRole', 'MonthlyIncome', 'Age', 'YearsWithCurrManager',
    'StockOptionLevel', 'YearsAtCompany'
]

# 'Attrition'을 숫자로 변환 (True=1, False=0)
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x else 0)

# 정규화 실행 (범주형 변수는 이 단계에서 제외하고 처리해야 할 수 있습니다)
df_normalized = pd.DataFrame(scaler.fit_transform(df[factors + ['Attrition']]), columns=factors + ['Attrition'])

# 퇴직 여부에 따른 평균 계산
attrition_means = df_normalized[df_normalized['Attrition'] == 1][factors].mean()
non_attrition_means = df_normalized[df_normalized['Attrition'] == 0][factors].mean()

# 시각화
fig = go.Figure()
fig.add_trace(go.Bar(x=factors, y=attrition_means, name="Attrition"))
fig.add_trace(go.Bar(x=factors, y=non_attrition_means, name="Non-Attrition"))

fig.update_layout(
    title="Normalized Values of Top Factors by Attrition Status",
    xaxis_title="Factors",
    yaxis_title="Normalized Average Value",
    barmode='group'
)

fig.show()
````
